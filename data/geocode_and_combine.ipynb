```python
from googlemaps import Client as GoogleMaps
import pandas as pd 
```

```python
gmaps = GoogleMaps('API_Key_Here')
```

```python
lat_longs = pd.read_csv("lat_longs.csv")
lat_longs.head()
```

```python
lat_longs['long'] = ""
lat_longs['lat'] = ""
```

```python
for x in range(len(lat_longs)):
    geocode_result = gmaps.geocode(lat_longs['full_address'][x])
    lat_longs['lat'][x] = geocode_result[0]['geometry']['location'] ['lat']
    lat_longs['long'][x] = geocode_result[0]['geometry']['location']['lng']
lat_longs.head()
```

```python
lat_longs.to_csv('with_lat_long.csv')
```
