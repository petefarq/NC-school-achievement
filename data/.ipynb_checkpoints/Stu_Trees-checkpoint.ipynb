```python
from sklearn import tree
import pandas as pd
import os
```

```python
df = pd.read_csv("main_data_tree.csv")
df.head(10)
```

```python
target = df["EOG_quintile"]
target_names = ["1", "2", "3", "4", "5"]
```

```python
data = df.drop("EOG_quintile", axis=1)
feature_names = ["percent_EDS","child_abuse_rate","children_conc_pov","elevated_lead","juv_delinquent","No_HSdegree","parent_unemployed","county_poc","county_tier","poverty_county","median_inc_county"]
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```
